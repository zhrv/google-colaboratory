```
import numpy as np
import matplotlib.pyplot as plt
import random as rnd

plt.figure(figsize=(10, 10))
plt.axis([-0.5,0.5,0,1])
_a = 60.0*np.pi/180.0
m_tree_p = [[np.cos(_a), -np.sin(_a)], 
            [np.sin(_a),  np.cos(_a)]]
_a = -60.0*np.pi/180.0
m_tree_m = [[np.cos(_a), -np.sin(_a)], 
            [np.sin(_a),  np.cos(_a)]]


def tree(n, p1, p2, ksi):
  p3 = [(1.0-ksi)*p1[0]+ksi*p2[0], (1.0-ksi)*p1[1]+ksi*p2[1]]
  ksi2 = ksi/2
  p6 = [(1.0-ksi2)*p1[0]+ksi2*p2[0], (1.0-ksi2)*p1[1]+ksi2*p2[1]]
  if (n <= 0):
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]])
  else:
    p4 = [m_tree_p[0][0]*(p2[0]-p3[0])+m_tree_p[0][1]*(p2[1]-p3[1])+p3[0], m_tree_p[1][0]*(p2[0]-p3[0])+m_tree_p[1][1]*(p2[1]-p3[1])+p3[1]]
    p5 = [m_tree_m[0][0]*(p2[0]-p3[0])+m_tree_m[0][1]*(p2[1]-p3[1])+p3[0], m_tree_m[1][0]*(p2[0]-p3[0])+m_tree_m[1][1]*(p2[1]-p3[1])+p3[1]]
    p7 = [m_tree_p[0][0]*(p3[0]-p6[0])+m_tree_p[0][1]*(p3[1]-p6[1])+p6[0], m_tree_p[1][0]*(p3[0]-p6[0])+m_tree_p[1][1]*(p3[1]-p6[1])+p6[1]]
    p8 = [m_tree_m[0][0]*(p3[0]-p6[0])+m_tree_m[0][1]*(p3[1]-p6[1])+p6[0], m_tree_m[1][0]*(p3[0]-p6[0])+m_tree_m[1][1]*(p3[1]-p6[1])+p6[1]]
    tree(0, p1, p3, ksi)
    tree(n-1, p3, p4, ksi)
    tree(n-1, p3, p5, ksi)
    tree(n-1, p3, p2, ksi)
    tree(n-1, p6, p7, ksi)
    tree(n-1, p6, p8, ksi)


tree(5, [0,0], [0,1], 0.66)
```

```
import numpy as np
import matplotlib.pyplot as plt
import random as rnd

plt.figure(figsize=(5, 10))
plt.axis([0,1,0,2])
angle = 60

_a = angle*np.pi/180.0
m = [[np.cos(_a), -np.sin(_a)], 
     [np.sin(_a),  np.cos(_a)]]

def koh(n, p1, p2):
  if (n == 0):
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]])
  else:
    p5 = [p1[0]/3.0+2.0*p2[0]/3.0, p1[1]/3.0+2.0*p2[1]/3.0]
    p3 = [2.0*p1[0]/3.0+p2[0]/3.0, 2.0*p1[1]/3.0+p2[1]/3.0]
    p4 = [m[0][0]*(p5[0]-p3[0])+m[0][1]*(p5[1]-p3[1])+p3[0], m[1][0]*(p5[0]-p3[0])+m[1][1]*(p5[1]-p3[1])+p3[1]]
    koh(n-1, p1, p3)
    koh(n-1, p5, p2)
    koh(n-1, p3, p4)
    koh(n-1, p4, p5)

level = 5

koh(level, [0,0], [1,0])

koh(level, [1,1], [0,1])
koh(level, [0.5,1.866], [1,1])
koh(level, [0,1], [0.5,1.866])
```

```
tri = [[0.0, 0.0], [0.5, 0.866], [1.0, 0.0]]
x = [0.0, 0.0]
#fig = plt.figure(figsize=(1,1))


def serpinsky(n):
  rnd.seed(0);
  

  for i in range(n):
    j = int(3.0*rnd.random())
    x[0] += tri[j][0]
    x[1] += tri[j][1]
    x[0] /= 2.0
    x[1] /= 2.0
    scatter1 = plt.scatter(x[0], x[1], marker='o', c='b', s=1)
```

```
import numpy as np
import matplotlib.pyplot as plt
import random as rnd

plt.figure(figsize=(6, 12))
plt.axis([-3,3,0,12])


def fern(n):
  rnd.seed(0);
  x1 = np.array([0.0, 0.0])
  x2 = np.array([0.0, 0.0])

  a1 = np.array([[  0.85, 0.04], 
                 [ -0.04, 0.85]])
  b1 = np.array([0, 1.6])
  
  a2 = np.array([[ -0.15, 0.28], 
                 [  0.26, 0.24]])
  b2 = np.array([0, 0.44])
  
  a3 = np.array([[  0.20, -0.26], 
                 [  0.23,  0.22]])
  b3 = np.array([0, 1.6])
  
  a4 = np.array([[  0.00,  0.00], 
                 [  0.00,  0.16]])
  b4 = np.array([0, 0])
  
  for i in range(n):
    k = int(100.0*rnd.random())
    if (k < 85):
      x2 = a1.dot(x1)+b1
    elif (k < 92):
      x2 = a2.dot(x1)+b2
    elif (k < 99):
      x2 = a3.dot(x1)+b3
    else:
      x2 = a4.dot(x1)+b4
      
    scatter1 = plt.scatter(x2[0], x2[1], marker='o', c='g', s=1)
    x1[0] = x2[0]
    x1[1] = x2[1]

    

fern(5000)
  
    

```

```
# Import libraries for simulation
import tensorflow as tf
import numpy as np

# Imports for visualization
import PIL.Image
from io import BytesIO
from IPython.display import Image, display

def DisplayFractal(a, fmt='jpeg'):
  """Display an array of iteration counts as a
     colorful picture of a fractal."""
  a_cyclic = (6.28*a/20.0).reshape(list(a.shape)+[1])
  img = np.concatenate([10+20*np.cos(a_cyclic),
                        30+50*np.sin(a_cyclic),
                        155-80*np.cos(a_cyclic)], 2)
  img[a==a.max()] = 0
  a = img
  a = np.uint8(np.clip(a, 0, 255))
  f = BytesIO()
  PIL.Image.fromarray(a).save(f, fmt)
  display(Image(data=f.getvalue()))

  
sess = tf.InteractiveSession()

# Use NumPy to create a 2D array of complex numbers

Y, X = np.mgrid[-1.3:1.3:0.005, -2:1:0.005]
Z = X+1j*Y

xs = tf.constant(Z.astype(np.complex64))
zs = tf.Variable(xs)
ns = tf.Variable(tf.zeros_like(xs, tf.float32))


tf.global_variables_initializer().run()

# Compute the new values of z: z^2 + x
zs_ = zs*zs + xs

# Have we diverged with this new value?
not_diverged = tf.abs(zs_) < 4

# Operation to update the zs and the iteration count.
#
# Note: We keep computing zs after they diverge! This
#       is very wasteful! There are better, if a little
#       less simple, ways to do this.
#
step = tf.group(
  zs.assign(zs_),
  ns.assign_add(tf.cast(not_diverged, tf.float32))
  )

for i in range(200): step.run()
  
DisplayFractal(ns.eval())
```
